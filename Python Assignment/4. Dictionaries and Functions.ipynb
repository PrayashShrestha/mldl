#### 1.1 Write a Python Program(with class concepts) to find the area of the triangle using the below
formula.
area = (s*(s-a)*(s-b)*(s-c)) * * 0.5

s = (a+b+c)/2
#### Function to take the length of the sides of triangle from user should be defined in the parent class and function to calculate the area should be defined in subclass.

```python
# Parent class to take input
class TriangleInp:
    def get_inp(self):
        self.a, self.b, self.c = [int(val) for val in input("Enter the sides of the triangle: ").split()] #taking multiple inputs using list comprehension
        

# Child class to print out the area of the triangle
class TriangleArea(TriangleInp):
    def calc_area(self):
        s=(self.a+ self.b+ self.c)/2
        area = ((s*(s- self.a)*(s- self.b)*(s- self.c)))
        return(round(abs(area) ** 0.5, 2))    
    
# Instantiate the Triangle class
if __name__ == "__main__":
    t = TriangleArea()
    t.get_inp() # Calling the function to take input
    print(f"The area of triangle is {t.calc_area()}") # Calling the function to give the area
    
```

### 1.2 Write a function filter_long_words() that takes a list of words and an integer n and returns the list of words that are longer than n.

```python
def filter_long_words(lst: list[str], n: int) -> list[str]:
    res = []
    for word in lst:
        if(len(word) > n):
            res.append(word)
    
    return res


    
```

```python
arr = ["hello", "hi", "namaste", "hey", "walekum"]
leng = 3

filter_long_words(lst = arr, n= leng)
```

### 2.2 Write a Python function which takes a character (i.e. a string of length 1) and returns True if it is a vowel, False otherwise.

```python
def vowel_checker(letter: str) -> bool:
    vowels = "aeiouAEIOU"
    
    if letter in vowels:
        return True
    else:
        return False

vowel_checker(input("Enter the characeter: "))
```
