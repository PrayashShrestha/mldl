#### 1. Write a Python Program to implement your own myreduce() function which works exactly like Python's built-in function reduce()

```python
# Using built in function result
from functools import reduce

lst = [2,5,9,6,3]

# multiplication function
def mul(a,b):
    return(a*b)

print(reduce(mul, lst))


```

```python
# creating my own reduce function

lst = [2,5,9,6,3]

# multiplication function
def mul(a,b):
    return a*b

# myreduce function
def myreduce(fun, lst):
    res = lst[0]
    for num in lst[1:]:
        res = fun(res, num)
    return res

print(myreduce(mul, lst))
```

#### 2. Write a Python program to implement your own myfilter() function which works exactly like python's built in function filter.

```python
lst = [10, 5, 22, 45, 88, 11, 9, 2, 6]
search = 11
```

##### Search using Linear Search

```python
def myfilter(num, arr):
    for i in range(0, len(arr)):
        if arr[i] == num:
            return True
    return False 

myfilter(search, lst)
```

Here the time complexity of Linear Search is O(n)

##### Search Using Binary Search

```python
def myfilter(num, arr):
    arr.sort()
    start = 0
    end = len(arr)-1

    while(start <= end):
        
        # Calculate the mid value
        mid = end + (start - end)//2  # just to prevent overflow otherwise we can use (start + end)//2
        
        # if the item is present return True
        if(arr[mid] == num):
            return True
        
        # include left part of array from the mid point
        if(arr[mid] > num):
             #print("Search to the left")
            end = mid - 1
            
        # include right part of array from mid point    
        else:
             #print("Search to the right")
            start = mid + 1
    return False

myfilter(search, lst)
```
